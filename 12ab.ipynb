I think I want to build a dictionary from the input rules. Determination of whether or not the sequence produces a plant will depend if the sequence is found in the library.

```python
# input data

initstate = '##.#...#.#.#....###.#.#....##.#...##.##.###..#.##.###..####.#..##..#.##..#.......####.#.#..#....##.#'
datain = np.loadtxt('12a_input.txt',comments=None,dtype=str,usecols=(0,2))#,delimiter=' ')

#test
initstate_test = '...#..#.#..##......###...###...........'
datain_test = np.loadtxt('12a_input_test.txt',comments=None,dtype=str,usecols=(0,2))#,delimiter=' ')
```

```python
#datain
```

```python
# create the dictionary

dic= {}
dic_test = {}

for row in datain:
    dic[row[0]] = row[1]
    
for row in datain_test:
    dic_test[row[0]] = row[1]
```

```python
startpot = -5
endpad = 150

def pot2idx(pot):
    idx = pot-startpot
    return idx

def idx2pot(idx):
    pot = idx + startpot
    return pot

def arr2str(array,frm,to):
    string = ''
    for i in range(frm,to):
        string += array[i][0]
    return string

def evolve(state):
    new = '..'
    for i in range(2,len(state)-2):
        new += dic[state[i-2:i+3]]
    new += '..'
    return new

def evolvetest(state):
    new = '..'
    for i in range(2,len(state)-2):
        if state[i-2:i+3] not in dic_test:
            new += '.'
        else:
            new += dic_test[state[i-2:i+3]]
    new += '..'
    return new
        
```

```python
# Pad the initial state
state = -1*startpot*'.' + initstate + endpad*'.'

#print(pots)
#print(state)
for loop in range(100):
    state = evolve(state)
    print(loop+1,' ',state[20:130])
```
